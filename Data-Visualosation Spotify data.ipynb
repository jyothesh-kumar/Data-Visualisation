## Data Quality Issues

The dataset may contain:
- Missing values
- Duplicate records
- Inconsistent column naming
- Mixed data types

These issues must be handled before visualization.

```python
songs_all = songs_all.drop_duplicates()

songs_all = songs_all.dropna()

songs_all.columns = songs_all.columns.str.lower().str.strip()
songs_2025.columns = songs_2025.columns.str.lower().str.strip()
artists_2025.columns = artists_2025.columns.str.lower().str.strip()
```

## 1: Dominant Artists



```python
top_artists = songs_all["artist"].value_counts().head(10)

plt.figure()
top_artists.plot(kind="bar", color="orange")
plt.title("Top 10 Artists (All-Time Songs)")
plt.xlabel("Artist")
plt.ylabel("Number of Songs")
plt.xticks(rotation=50)
plt.tight_layout()
plt.show()
plt.figure()
```

## 9:Average Audio Features: All-Time Songs vs 2025 Top Songs


```python
import matplotlib.pyplot as plt
import seaborn as sns

comparison = pd.DataFrame({
    'Dataset': ['All-Time Songs', '2025 Top Songs'],
    'Avg Danceability': [songs_all['danceability'].mean(), songs_2025['danceability'].mean()],
    'Avg Energy': [songs_all['energy'].mean(), songs_2025['energy'].mean()],
    'Avg Valence': [songs_all['valence'].mean(), songs_2025['valence'].mean()],
    'Avg Acousticness': [songs_all['acousticness'].mean(), songs_2025['acousticness'].mean()]
})

comparison_melted = comparison.melt(id_vars='Dataset', var_name='Feature', value_name='Average')

plt.figure(figsize=(10,5))
ax = sns.barplot(
    data=comparison_melted, 
    x='Feature', 
    y='Average', 
    hue='Dataset', 
    palette="Dark2"
)

plt.title("Average Audio Features: All-Time vs 2025 Songs", fontsize=14)
plt.xlabel("Audio Feature", fontsize=10)
plt.ylabel("Average Value", fontsize=10)
plt.xticks(rotation=20, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10, title='Dataset', title_fontsize=11)
plt.tight_layout() 
plt.show()
```

# Explanation

This grouped bar chart compares the average values of four important audio features:

Danceability
Energy
Valence
Acousticness

between:

All-Time Songs
2025 Top Songs

The chart helps identify how modern popular songs differ from historically successful songs.


# Insight

This visualization helps understand changes in music style and listener preferences over time.
